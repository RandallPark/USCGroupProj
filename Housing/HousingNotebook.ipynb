````python
#Create DataFrame with just 2016(most recent) data
housing_2016 = housing_data[housing_data.Date == '01/01/2016']

housing_2016 = housing_2016.dropna(subset=['Amount'])



#Reset index for new DF
housing_2016 = housing_2016.reset_index()

#Drop column created by old index
housing_2016 = housing_2016.drop('index', axis=1)

#Drop column not relevant.
housing_2016 = housing_2016.drop(['Policy Area','Year','Variable','Dataset', 'Tract', 'Row ID', 'Date', 'GEOID' ], axis=1)

#Rename column
housing_2016= housing_2016.rename(columns = {'Amount': 'Median Rent 2016'})



housing_2016
#by_neighborhood = housing_data.goupby
````
